### Relu Activation function

```python
import matplotlib.pyplot as plt

def relu(x):
    return max(0.0,x)

inputs = [i for i in range(-15,15)]
outputs = [relu(i) for i in inputs]
plt.plot(inputs, outputs) 
plt.show()
```

### Sigmoid Activation function

```python
import matplotlib.pyplot as plt
from math import exp
def sigmoid(x):
    return 1.0/(1.0 + exp(-x))

inputs = [i for i in range(-15,15)]
outputs = [sigmoid(i) for i in inputs]
plt.plot(inputs, outputs) 
plt.show()
```

### softmax activation function

```python
import numpy as np
from math import exp
def softmax(x):
    sum_exp = 0
    sum = 0
    y = []
    for i in x:
        sum_exp += exp(i)
    for i in x:
        y.append(exp(i)/sum_exp)
        sum+=exp(i)/sum_exp
    return y
    
inputs  = np.array([i for i in range(-15,15)])
print(inputs.ndim)
print(inputs.size)
print(inputs.shape)
print(softmax(inputs))
```
