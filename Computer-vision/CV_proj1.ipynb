```python
import rasterio
import matplotlib.pyplot as plt
import numpy as np
```

### Analysing a single image for the Coimbatore Satellite images over a single year

```python
# Open the file
with rasterio.open('NDVI_Coimbatore.tif') as src:
    ndvi = src.read(1)  # read the first band
    num_bands = src.count  # get the number of bands

# Display the image
plt.imshow(ndvi, cmap='RdYlGn')
plt.colorbar()
plt.title('NDVI Coimbatore')
plt.show()
print(f'Number of bands: {num_bands}')
```

```python
ndvi.shape
```

```python
mean_ndvi = np.mean(ndvi)
median_ndvi = np.median(ndvi)
std_dev_ndvi = np.std(ndvi)

# Print the statistics
print('Mean NDVI:', mean_ndvi)
print('Median NDVI:', median_ndvi)
print('Standard Deviation of NDVI:', std_dev_ndvi)
```

```python
import cv2
import numpy as np

# Normalize NDVI values to the range [0, 255]
ndvi_normalized = ((ndvi - ndvi.min()) / (ndvi.max() - ndvi.min()) * 255).astype(np.uint8)

# Apply Canny edge detection
edges = cv2.Canny(ndvi_normalized, 50, 100)

# Apply Harris corner detection
dst = cv2.cornerHarris(ndvi.astype(np.float32), 2, 3, 0.04)

# Create a boolean bitmap of corner positions
corners = dst > 0.01 * dst.max()

# Find the coordinates from the boolean bitmap
coord = np.argwhere(corners)

# Draw circles on the coordinates to mark the corners
ndvi_colored = cv2.cvtColor(ndvi_normalized, cv2.COLOR_GRAY2BGR)
for y, x in coord:
    cv2.circle(ndvi_colored, (x, y), 5, (0, 0, 255), 2)

# Display the results
plt.figure(figsize=(12, 6))


plt.subplot(1, 3, 1)
plt.imshow(ndvi, cmap='RdYlGn')
plt.title('Original NDVI Image')


plt.subplot(1, 3, 2)
plt.imshow(ndvi_colored)
plt.title('NDVI with Corners')

plt.subplot(1, 3, 3)
plt.imshow(edges, cmap='gray')
plt.title('Canny Edge Detection')



plt.show()
```
