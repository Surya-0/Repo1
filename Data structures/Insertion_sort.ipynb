```python
import time
import random
import matplotlib.pyplot as plt
random.seed = 42

def insertion_sort(arr):
    for i in range(1,len(arr)):
        key = arr[i]
        j= i-1
        while j>=0 and key<arr[j]:
            arr[j+1]=arr[j]
            j-=1
        arr[j+1]=key

m = 5
size_arr = []
execution_time = []
for i in range(0,m):
    start_time = time.time()
    n = random.randint(10,100)
    print(n)
    size_arr.append(n)
    arr = []
    for i in range(n):
        arr.append(random.randint(10,100))
    # print("The random array is : ",arr)
    insertion_sort(arr)
    # print("The sorted array is : ",arr)
    end_time = time.time()
    taken_time = end_time - start_time
    execution_time.append(taken_time)

print(execution_time)
plt.scatter(size_arr,execution_time,marker="*")
plt.plot(size_arr,execution_time)
plt.xlabel("Size of array")
plt.ylabel("Execution Time")
plt.show()
```
