```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
df = pd.read_csv('Mall_Customers.csv')
df
```

```python
X = df.iloc[:,3:].values
```

```python
X
```

```python
wcss_list = []
for i in range(1, 11):
    kmeans =KMeans(n_clusters=i, init='k-means++', random_state= 42)
    kmeans.fit(X)
    wcss_list.append(kmeans.inertia_)
plt.plot(range(1, 11), wcss_list)
plt.title('The Elobw Method Graph')
plt.xlabel('Number of clusters(k)')
plt.ylabel('wcss_list')
plt.show()
```

```python
len(wcss_list)
```

```python
kmeans = KMeans(n_clusters= 5,random_state=42)
kmeans.fit(X)
```

```python
y_pred = kmeans.predict(X)
```

```python
y_pred
```

```python
arr = [0,0,0,0,0]
for i in range(len(y_pred)):
    if y_pred[i] == 0:
        arr[0]+=1
    elif y_pred[i] == 1:
        arr[1]+=1
    elif y_pred[i] == 2:
        arr[2]+=1
    elif y_pred[i] == 3:
        arr[3]+=1
    elif y_pred[i] == 4:
        arr[4]+=1
```

```python
plt.scatter(X[y_pred == 0,0],X[y_pred == 0,1],s = 100,c = 'blue',label = 'Cluster 1')
plt.scatter(X[y_pred == 1,0],X[y_pred == 1,1],s = 100,c = 'green',label = 'Cluster 2')
plt.scatter(X[y_pred == 2,0],X[y_pred == 2,1],s = 100,c = 'yellow',label = 'Cluster 3')
plt.scatter(X[y_pred == 3,0],X[y_pred == 3,1],s = 100,c = 'silver',label = 'Cluster 4')
plt.scatter(X[y_pred == 4,0],X[y_pred == 4,1],s = 100,c = 'aqua',label = 'Cluster 5')
plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],s = 300,c = 'black',label = 'Centroid')
plt.legend()
plt.show()
```
