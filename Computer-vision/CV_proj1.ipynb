## This notebook is primarily for experimentation

```python
import rasterio
import matplotlib.pyplot as plt
import numpy as np
```

### Analysing a single image for the Coimbatore Satellite images over a single year

```python
# Open the file
with rasterio.open('NDVI_Coimbatore.tif') as src:
    ndvi = src.read(1)  # read the first band
    num_bands = src.count  # get the number of bands

# Display the image
plt.imshow(ndvi, cmap='RdYlGn')
plt.colorbar()
plt.title('NDVI Coimbatore')
plt.show()
print(f'Number of bands: {num_bands}')
```

```python
mean_ndvi = np.mean(ndvi)
median_ndvi = np.median(ndvi)
std_dev_ndvi = np.std(ndvi)

# Print the statistics
print('Mean NDVI:', mean_ndvi)
print('Median NDVI:', median_ndvi)
print('Standard Deviation of NDVI:', std_dev_ndvi)
```
