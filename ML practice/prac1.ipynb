```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as mpl
import seaborn as sns
```

```python
data = [['apple',10,240],['orange',15,150],['grapes',14,80],['pineapple',10,70]]
df1 = pd.DataFrame(data,columns=['Fruit_name','Quantity','Unit_price'])
df1
```

```python
quant = []
quant = df1['Quantity']
quant = np.array(quant)
Fruits = []
Fruits = df1['Fruit_name']
Fruits = np.array(Fruits)
print(Fruits)
print(quant)
```

```python
df1.describe()
```

```python
mpl.bar(Fruits,quant)
mpl.xlabel('Fruits')
mpl.ylabel('Unit_price')
mpl.title('Bar plot example')
mpl.show()
```
