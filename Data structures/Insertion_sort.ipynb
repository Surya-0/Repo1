```python
import heapq

def merge_k_sorted_arrays(arrays):
    # Create a min-heap to store the smallest element from each subarray
    min_heap = []
    
    # Initialize the result array
    result = []
    
    # Populate the min-heap with the first element from each subarray
    for i, arr in enumerate(arrays):
        if arr:
            heapq.heappush(min_heap, (arr[0], i, 0))
    
    # Merge the subarrays until the min-heap is empty
    while min_heap:
        val, subarray_index, index_in_subarray = heapq.heappop(min_heap)
        print(val , subarray_index , index_in_subarray)
        # Add the smallest element to the result
        result.append(val)
        
        # Check if there are more elements in the current subarray
        if index_in_subarray + 1 < len(arrays[subarray_index]):
            next_val = arrays[subarray_index][index_in_subarray + 1]
            heapq.heappush(min_heap, (next_val, subarray_index, index_in_subarray + 1))
    
    return result

# Example usage
k_sorted_arrays = [[1, 3, 5], [2, 4, 6], [0, 7, 8]]
sorted_array = merge_k_sorted_arrays(k_sorted_arrays)
print(sorted_array)
```

```python
def split_array(array,k):
    n = len(array)
    sub_array_size = n//k
    rem_elements = n%k
    start_index = 0
    sub_array = []
    for i in range (k):
        sub_array_end = start_index + sub_array_size
        sub_array.append(array[start_index:sub_array_end])
        start_index = sub_array_end
        
    if rem_elements > 0:
        sub_array[-1].extend(array[sub_array_end:])
    return sub_array
        
```
